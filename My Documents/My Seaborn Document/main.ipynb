```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
tips
```

#  Scatter Plot

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size')
```

# Histogram

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

# Bar Plot

```python
sns.barplot(x='sex', y='tip', data=tips, hue='sex')
```

# Box Plot

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

# Strip Plot

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', dodge=True)
```

# Join Plot

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='YlGnBu')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex', cmap='YlGnBu')
```

# Pair Plot

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

# Heat Map

```python
titanic = titanic.drop(columns=['sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'])
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu')
```

# Cluster Map

```python
sns.clustermap(iris.drop('species', axis=1))
```
